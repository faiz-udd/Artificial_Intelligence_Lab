```python
import numpy 

values=[23,45,6,7,8,9,24,10,11,15,12,8,8,8]

print(values)
x = numpy.mean(values)
print(x)
```

```python
x1 = numpy.median(values)
print(x1)
```

```python
from scipy import stats

#speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]


x2 = stats.mode(values)

print(x2)
```

```python
val1 =numpy.array([[1,2,3],[3,2,1],[2,1,3]])
val2 =numpy.array([1,2,3])
val3 =numpy.array([[1],[2],[3]])
print(val1)
print(val2)
print(val3)
```

```python
# standard deviation

x3 = numpy.std(values)
print(x3)

array1 = [-3,-2,-1,0,1,2,3]
mean1 =numpy.mean(array1)
print(mean1)

std1 = numpy.std(array1)
print(std1)
```

```python
#percentile Formuala: Percentile is found with the equation: P = n/N * 100%. Where P is the percentile, lower case n is the number of data points below the data point of interest, and N is the total number of data points in the data set.
x4 = numpy.percentile(values, 25)
print(x4)
```
